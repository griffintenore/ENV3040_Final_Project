```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

file = open("final_project_data.csv") #Spreadsheet csv containing 6 variables from the Everglades dataset

list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []


#Extract data from columns and create lists of data points:

first = True

for line in file:
    if first == True:
        data = line.split(",")
        first = False
    else:
        data = line.split(",")
        list1.append(data[1])
        list2.append(data[2])
        list3.append(data[3])
        list4.append(data[4])
        list5.append(data[5])
        list6.append(data[6])

#Convert list elements from strings to floats:         

pH_data = [float(x) for x in list1]
chlorophyll_data = [float(x) for x in list2]
sulfate_data = [float(x) for x in list3]
carbon_data = [float(x) for x in list4]
oxygen_data = [float(x) for x in list5]
temp_data = [float(x) for x in list6]


#Mean, median, and standard deviation calculation:

mean_pH = np.mean(pH_data)
median_pH = np.median(pH_data)
stdev_pH = np.std(pH_data)

mean_chlor = np.mean(chlorophyll_data)
median_chlor = np.median(chlorophyll_data)
stdev_chlor = np.std(chlorophyll_data)

print("The mean pH reading is: {:0.2f}\n".format(mean_pH),"The median pH reading is:", median_pH, "\nThe standard deviation in the pH data is: {:0.2f}".format(stdev_pH))
print("\nThe mean chlorophyll concentration is: {:0.2f} ug/L\n".format(mean_chlor),"The median chlorophyll concentration is:", median_chlor, "ug/L\nThe standard deviation in the chlorophyll data is: {:0.2f} ug/L".format(stdev_chlor))

pH_array = np.array(pH_data)
chlor_array = np.array(chlorophyll_data)


sulfate_array = np.array(sulfate_data)
carbon_array = np.array(carbon_data)
oxygen_array = np.array(oxygen_data)
temp_array = np.array(temp_data)
```

```python
#pH Histogram

import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

plt.xlabel("pH")
plt.ylabel("Frequency")
plt.hist(pH_array, bins = 10)
plt.title("pH Reading Frequency")
```

```python
#Chlorophyll histogram

import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

plt.xlabel("Chlorophyll Concentration (ug/L)")
plt.ylabel("Frequency")
plt.hist(chlor_array, bins = 10)
plt.hist(chlor_array, color = "red")
plt.title("Chlorophyll Concentration Frequency")
```

```python
#Pearson and Spearman calculations

pearson = stats.pearsonr(pH_array, chlor_array)
print("The Pearson correlation coefficent is: {:.3f}".format(pearson.statistic), "\nThe Pearson p-value is: {:0.2e}".format(pearson.pvalue))

spearman = stats.spearmanr(pH_array, chlor_array)
print("\nThe Spearman correlation coefficent is: {:.2e}".format(spearman.statistic), "\nThe Spearman p-value is: {:0.2e}".format(spearman.pvalue))


#Scatter plots and linear regression
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress


fig, ax = plt.subplots()
ax.scatter(pH_array, chlor_array)
ax.set_xlabel("pH")
ax.set_ylabel("Chlorophyll Concentration (ug/L)")
plt.title("Chlorophyll Concentration vs. pH")
#ax.set_xlim(0, 14) 

fitline = np.polyfit(pH_array, chlor_array, 1)


p = np.poly1d(fitline)
plt.plot(pH_array, p(pH_array), "red")

slope, intercept = fitline

print("The line of best fit can be exressed as: y = {:0.2f}x + ".format(slope),"{:0.2f}".format(intercept)) 
```

```python
#PCA
import pandas as pd
import sys
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.decomposition import PCA
from sklearn import preprocessing


LoL = []
LoL.append(pH_data)
LoL.append(chlorophyll_data)
LoL.append(sulfate_data)
LoL.append(carbon_data)
LoL.append(oxygen_data)
LoL.append(temp_data)


data = np.array(LoL)

#Format data for PCA

scaled_data = preprocessing.scale(data.T)
pca = PCA()
pca.fit(scaled_data)
pca_data = pca.transform(scaled_data)


principal_components = pca.fit_transform(data)
print(principal_components)
print(pca.explained_variance_ratio_)

#Create a bar graph 

per_var = np.round(pca.explained_variance_ratio_*100, decimals =1)
labels = ['PC' +str(x) for x in range(1, len(per_var)+1)]
plt.bar(x=range(1, len(per_var)+1), height = per_var, tick_label = labels)
plt.ylabel("Percentage of Explained Variance")
plt.xlabel("Principal Component")
plt.title("Scree Plot")
```
